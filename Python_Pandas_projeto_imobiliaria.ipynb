Para % de cada tipo de imóvel e valor médio de venda de cada imóvel serão necessárias as colunas Tipo e Valor

```python
df_preco_tipo = dados.groupby("Tipo", sort = True)[["Valor"]].mean()
df_preco_tipo.plot(kind = "barh", figsize= (10,10), color="purple")
```

# Manipulando os dados

## Criando colunas numéricas

```python
url =  'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep = ';')
dados
```

```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
dados['Valor_por_ano'] = (dados['Valor'] + dados['Condominio'])*12 + dados['IPTU']
dados.head()
```

## Criando colunas categóricas

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com '+ \
                    dados['Quartos'].astype(str) + ' quarto(s) e ' + \
                    dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head()
```

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
dados.head()
```

```python
dados.to_csv('dados_completos_dev.csv', index = False, sep = ';')
```
